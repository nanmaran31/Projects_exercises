```python
def formation_class(x):
    if x == '3-4-3':
        formation_class = 343
    elif x == '4-3-3':
        formation_class = 433
    elif x == '4-2-3-1':
        formation_class = 4231
    elif x == '3-5-1-1':
        formation_class = 4231
    elif x == '4-1-4-1':
        formation_class = 4141
    elif x == '3-5-2':
        formation_class = 352
    elif x == '4-2-2-2':
        formation_class = 4222
    elif x == '3-4-1-2':
        formation_class = 3412
    elif x == '4-3-2-1':
        formation_class = 4312
        
    return formation_class
```

```python
df['formation_class'] = df['Formation'].apply(formation_class)
```

```python
df['formation_class'].value_counts()
```

```python
def win_nowin(x):
    if x == 'W':
        win_nowin = 1
    else:
        win_nowin = 0
    return win_nowin
```

```python
df['win_nowin'] = df['Result'].apply(win_nowin)
```

```python
df['win_nowin'].value_counts()
```

```python
df['win_nowin'].hist(figsize = (5,5))
plt.xlabel('Result', color ='White')
plt.ylabel('Count', color ='White')
plt.title('Win and No-Win Counts (Target Variable)', color ='White')
plt.legend(['0 = No-Win \n 1 = Win '],bbox_to_anchor = (1.05, 0.6))
plt.savefig('target.png')
```
